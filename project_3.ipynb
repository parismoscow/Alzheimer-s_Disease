```python
df = pd.DataFrame(
    selected_file_one
)
df.head()
```

```python
def V_pct_change(df):
    df['V_pct'] = 100 * (1 - df.iloc[0].Ventricles / df.Ventricles)
    return df

df=df.groupby('PTID').apply(V_pct_change)
```

```python
def H_pct_change(df):
    df['H_pct'] = 100 * (1 - df.iloc[0].Hippocampus / df.Hippocampus)
    return df

df=df.groupby('PTID').apply(H_pct_change)
df.head()
```

```python
def Wb_pct_change(df):
    df['Wb_pct'] = 100 * (1 - df.iloc[0].WholeBrain / df.WholeBrain)
    return df

df=df.groupby('PTID').apply(Wb_pct_change)
df.head()
```

```python
def E_pct_change(df):
    df['E_pct'] = 100 * (1 - df.iloc[0].Entorhinal / df.Entorhinal)
    return df

df=df.groupby('PTID').apply(E_pct_change)
df.head()
```

```python
def Mt_pct_change(df):
    df['Mt_pct'] = 100 * (1 - df.iloc[0].MidTemp / df.MidTemp)
    return df

df=df.groupby('PTID').apply(Mt_pct_change)
df.head()
```

```python
df.to_csv("pct_changes.csv", index=False, header=True)
```
